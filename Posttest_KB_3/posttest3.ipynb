## Grafik Boxplot

Grafik Boxplot merupakan ringkasan distribusi sampel yang disajikan secara grafis yang bisa menggambarkan bentuk distribusi data (skewness), ukuran tendensi sentral dan ukuran penyebaran (keragaman) data pengamatan.

```python
plt.boxplot(payment[['payment_installments']], labels=['Cash'])
plt.title('GRAFIK BOXPLOT')
plt.xlabel('Jenis Pembayaran')
plt.ylabel('JUmlah')
plt.show()
```

## Grafik Heatmap
Heatmap disebut juga visualisasi atau pemetaan dengan menampilkan data dengan cara representasi warna yang berbeda-beda. Biasanya pada heatmap semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap dan umumnya disimbolkan dengan warna berwarna biru. Grafik ini biasa digunakan untuk menunjukkan korelasi antar variable.

```python
sns.heatmap(payment.corr(), cmap ='Blues')
plt.title('GRAFIK HEATMAP')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

## Diagram garis jumlah orderan perbulan tahun 2016 - 2018

```python
df['order_approved_at'] = pd.to_datetime(df.order_approved_at, format='%Y-%m-%d')

order_monthly = df[['order_approved_at']].groupby(df['order_approved_at'].dt.to_period('M')).agg({'order_approved_at':'count'})
order_monthly = order_monthly.rename(columns = {'order_approved_at': 'num_of_orders'})
order_monthly = order_monthly.reset_index()

#menambahkan kolom untuk nama bulan dan tahun 
order_monthly['month_year'] = order_monthly['order_approved_at'].dt.strftime('%b-%Y')
order_monthly.head()
```

```python
plt.figure(figsize=(17,5))
sns.set_theme(style="ticks", font_scale=0.7)
order_monthly_viz = sns.lineplot(x = 'month_year', y = 'num_of_orders', data = order_monthly)
order_monthly_viz.set_title('Angka Order perbulan Tahun 2016-2018',fontsize = 20)
order_monthly_viz.set_xlabel('Bulan-Tahun',fontsize = 14)
order_monthly_viz.set_ylabel('Angka Order',fontsize = 14)
plt.show()
```

Dapat kita lihat, jumlah orderan tertinggi berada pada bulan november 2017 periode oktober 2016 hingga agustus 2018
